```python
import pandas as pd

nome_colunas = {
    'RAZÃO SOCIAL': 'razao_social',
    'ENDEREÇO': 'endereco',
    'BAIRRO': 'bairro',
    'BANDEIRA': 'bandeira',
    'PREÇO VENDA': 'preco',
    'DATA COLETA': 'data'
}

# Ajustar as planilhas para ter somente os dados

df_aditivada = pd.read_excel('BH_Aditivada.xlsx').rename(columns=nome_colunas)
df_aditivada['combustivel'] = 'GA'
```

```python
# Checar se o nome das colunas está certo, ajustar com o parâmetro header
df_comum = pd.read_excel('BH_Comum.xlsx').rename(columns=nome_colunas)
df_comum['combustivel'] = 'GC'
```

```python
# Junta os dois dataframes em um só
df = pd.concat([df_aditivada, df_comum]).reset_index(drop=True)
df.to_excel('data.xlsx')
```

```python
# 1)
print(f'''
GA: Desvio padrão: {df_aditivada.preco.std()} Média: {df_aditivada.preco.mean()}
GC: Desvio padrão: {df_comum.preco.std()} Média: {df_comum.preco.mean()}
''')
```

```python
# 2)
import seaborn as sns

sns.set_theme(style="whitegrid", palette="muted")

ax = sns.swarmplot(data=df, x="preco", y="combustivel")
ax.set(ylabel="")
```
